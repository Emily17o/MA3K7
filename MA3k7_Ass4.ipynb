```python
import math
import matplotlib.pyplot as plt

def calculate_P(m):
    floor_m_half = math.floor(m / 2)
    return sum(math.comb(i, 2*i - (m-1)) * (0.5)**i for i in range(floor_m_half, m))

# Input the upper limit for m
upper_limit = int(input("Enter the upper limit for m: "))

# Calculate P(m) for different values of m
m_values = list(range(1, upper_limit + 1))
P_values = [calculate_P(m) for m in m_values]

# Plot the graph
plt.plot(m_values, P_values, marker='o')
plt.title('Graph of P(m)')
plt.xlabel('m')
plt.ylabel('P(m)')
plt.grid(True)
plt.show()
```
