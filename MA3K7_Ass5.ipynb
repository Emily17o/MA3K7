```python
import math
import matplotlib.pyplot as plt

def calculate_a_m(m, a_prev):
    """
    Calculate the value of a_m based on the formula a_m = a_{m-1} + ceil(m/2)

    Parameters:
        m (int): Current index value
        a_prev (int): Value of a_{m-1}

    Returns:
        int: Value of a_m
    """
    return a_prev + math.ceil(m / 2)

# Example usage:
a_0 = 1  # Initial value of a_0
m = 15   # Value of m
a_values = [a_0]
for i in range(1, m):
    a_values.append(calculate_a_m(i, a_values[-1]))

# Print the result
print(f"a_{m} = {a_values[-1]}")

# Plotting the values of a_m
plt.plot(range(1, m+1), a_values, marker='o')
plt.title('Values of $a_m$ vs. $m$')
plt.xlabel('$m$')
plt.ylabel('$a_m$')
plt.grid(True)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

def calculate_a_m(m, a_prev):
    """
    Calculate the value of a_m based on the formula a_m = a_{m-1} + ceil(m/2)

    Parameters:
        m (int): Current index value
        a_prev (int): Value of a_{m-1}

    Returns:
        int: Value of a_m
    """
    return a_prev + math.ceil(m / 2)

# Example usage:
a_0 = 1  # Initial value of a_0
m = 21   # Value of m
a_values = [a_0]
for i in range(1, m):
    a_values.append(calculate_a_m(i, a_values[-1]))

# Print the result
print(f"a_{m} = {a_values[-1]}")

# Plotting the values of a_m starting from m=11
start_m = 11
plt.plot(range(start_m, m+1), a_values[start_m-1:], marker='o')
plt.title('Values of $a_m$ vs. $m$')
plt.xlabel('$m$')
plt.ylabel('$a_m$')
plt.grid(True)

# Adding horizontal lines y = 6b + 1
for b in range(4, 19):  # Adjust range as needed
    y_values = [6*b + 1] * (m - start_m + 1)
    plt.plot(range(start_m, m+1), y_values, linestyle='--', color='gray')
    plt.text(m + 0.5, 6*b + 1, f'b = {b}', verticalalignment='center', color='gray')

plt.show()

```
