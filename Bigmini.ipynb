```python
import math
import matplotlib.pyplot as plt

def binomial_coefficient(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

def calculate_sum(S):
    return sum(i / binomial_coefficient(2 * i, 2) for i in range(1, S + 1))

def plot_sum(max_S):
    S_values = list(range(1, max_S + 1))
    sum_results = [calculate_sum(S) for S in S_values]

    plt.plot(S_values, sum_results)
    plt.xlabel('Number of strings')
    plt.ylabel('Average Number of loops')
    plt.title('Average Number of loops vs Number of strings')
    plt.grid(True)
    plt.show()

# Example usage:
max_S = 500
plot_sum(max_S)
```

```python
def partition_count(n):
    # Initialize a list to store the partition counts
    partitions = [0] * (n + 1)
    partitions[0] = 1  # Base case: there is 1 way to partition 0

    # Iterate through numbers from 1 to n
    for i in range(1, n + 1):
        # Iterate through partition sizes
        for j in range(i, n + 1):
            # Update the partition count for each partition size
            partitions[j] += partitions[j - i]

    # Return the number of partitions for n
    return partitions[n]

# Example usage:
n = 4
print("Number of partitions for", n, "is:", partition_count(n))
```
